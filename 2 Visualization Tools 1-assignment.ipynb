## Assignment

Import necessary library

```python
import numpy as np  # useful for many scientific computing in Python
import pandas as pd # primary data structure library
```

Read the data

```python
df_can = pd.read_excel('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DV0101EN/labs/Data_Files/Canada.xlsx',
                       sheet_name='Canada by Citizenship',
                       skiprows=range(20),
                       skipfooter=2)
```

Drop unnecessary columns

```python
# in pandas axis=0 represents rows (default) and axis=1 represents columns.
df_can.drop(['AREA','REG','DEV','Type','Coverage'], axis=1, inplace=True)
df_can.head(2)
```

Rename columns title

```python
df_can.rename(columns={'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace=True)
df_can.columns
```

Add a 'Total' column

```python
df_can['Total'] = df_can.sum(axis=1)
df_can.head(40)
```

**Question 1:** Let's compare the number of immigrants from India and China from 1980 to 2013.


Step 1: Get the data set for China and India, and display dataframe.

```python
### type your answer here
df_can.set_index('Country', inplace=True)
df_can.index.name = None
df_can
```

```python
years = [i for i in range(1980,2014)]
china = df_can.loc['China',years]
india = df_can.loc['India', years]

print(china)
print(india)
```

Step 2: Plot graph. We will explicitly specify line plot by passing in `kind` parameter to `plot()`.

```python
### type your answer here
import matplotlib.pyplot as plt 

china.index = china.index.map(int) # let's change the index values of Haiti to type integer for plotting
china.plot(kind='line', label="China")

india.index = india.index.map(int) # let's change the index values of Haiti to type integer for plotting
india.plot(kind='line', label="India")

plt.legend()
plt.title('Imigran dari India dan China')
plt.ylabel('Jumlah Imigran')
plt.xlabel('Tahun')
plt.show()

```

**Question 2:** Compare the trend of top 5 countries that contributed the most to immigration to Canada.

Step 1: Get the data set for top 5 countries

```python
### type your answer here

df_can_sorted = df_can.sort_values(by="Total",ascending=False)
df_5_negara = df_can_sorted.head(5)

```

Step 2: Plot graph

```python
### type your answer here

x = df_5_negara.index
y = df_5_negara['Total']

plt.bar(x, y)
plt.title('Total Imigran ke Kanada (1980-2013)')
plt.ylabel('Jumlah Imigran')
plt.xticks(rotation=90)
plt.show()
```

**Question 3**: Create an unstacked area plot of the 5 countries that contributed the least to immigration to Canada **from** 1980 to 2013. Use a transparency value of 0.55.

```python
### type your answer here

df_bottom5 = df_can.sort_values(by="Total",ascending=True).head(5)

df_bottom5 = df_bottom5[years].transpose()

df_bottom5.plot.area(alpha=0.55)

plt.title('Imigrasi Terendah ke Kanada (1980-2013)')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Imigran')

plt.show()



```

**Question 4**: Display the immigration distribution for Greece, Albania, and Bulgaria for years 1980 - 2013? Use an overlapping plot with 15 bins and a transparency value of 0.35.

```python
### type your answer here

df_subset = df_can.loc[['Greece', 'Albania', 'Bulgaria'], years]
df_subset = df_subset.transpose()

# Membuat histogram dengan 15 bins dan transparansi 0.35
df_subset.plot(kind='hist', bins=15, alpha=0.35, stacked=True)


plt.title('Distribusi Imigrasi ke Kanada dari Yunani, Albania, dan Bulgaria (1980-2013)')
plt.xlabel('Jumlah Imigran')
plt.ylabel('Jumlah Tahun')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

plt.show()



```

**Question 5:** Create a *horizontal* bar plot showing the *total* number of immigrants to Canada from the top 15 countries, for the period 1980 - 2013. Label each country with the total immigrant count.

Step 1: Get the data pertaining to the top 15 countries.

```python
### type your answer here

df_upper15 = df_can.sort_values(by="Total",ascending=False).head(15)
df_upper15


```

Step 2: Plot data:
   1. Use `kind='barh'` to generate a bar chart with horizontal bars.
   2. Make sure to choose a good size for the plot and to label your axes and to give the plot a title.
<!--    3. Loop through the countries and annotate the immigrant population using the anotate function of the scripting interface. -->

```python
### type your answer here


plt.figure(figsize=(10, 8))
plt.barh(df_upper15.index, df_upper15['Total'], color='steelblue')

# Menambahkan label untuk setiap bar dengan jumlah total imigran
for index, value in enumerate(df_upper15['Total']):
    plt.text(value, index, str(value))

plt.title('Top 15 Negara Imigran ke Kanada (1980 - 2013)')
plt.xlabel('Total Imigran')
plt.ylabel('Negara')

plt.show()
```
