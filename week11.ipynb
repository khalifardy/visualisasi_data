```python
import networkx as nx
import matplotlib.pyplot as plt

lingko = nx.Graph()
daftar_kota = [
    'Bekasi',
    'Bandung',
    'Bogor',
    'Depok',
    'Sukabumi',
    'Cianjur',
    'Karawang',
    'Purwakarta',
    'Cimahi',
    'Garut',
    'Sumedang',
    'Tasikmalaya',
    'Cirebon',
    'Kuningan',
    'Subang',
    'Indramayu',
    'Majalengka',
    'Ciamis',
    'Banjar'
]
lingko.add_nodes_from(daftar_kota)
lingko.nodes()
nx.draw(lingko, with_labels=True)
plt.show()
```

```python
def edge_connection(kota_a, list_tetangga):
    for kota in daftar_kota:
        if kota in list_tetangga:
            lingko.add_edge(kota_a,kota)
            lingko.edges()
```

```python
edge_connection("Bekasi",['Karawang','Bogor','Depok'])
edge_connection("Depok",['Bogor'])
edge_connection("Bogor",['Sukabumi','Cianjur','Karawang'])
edge_connection("Sukabumi",['Cianjur'])
edge_connection("Cianjur",['Bandung','Garut','Karawang','Purwakarta'])
edge_connection("Karawang",['Purwakarta','Subang'])
edge_connection("Purwakarta",['Bandung','Subang'])
edge_connection("Bandung",['Garut','Subang','Sumedang','Cimahi'])
edge_connection("Garut",['Tasikmalaya','Sumedang','Majalengka'])
edge_connection("Sumedang",['Majalengka','Subang','Indramayu'])
edge_connection("Subang",['Indramayu'])
edge_connection("Indramayu",['Majalengka','Cirebon'])
edge_connection("Majalengka",['Cirebon','Kuningan','Ciamis','Tasikmalaya'])
edge_connection("Tasikmalaya",['Ciamis'])
edge_connection("Ciamis",['Kuningan','Banjar'])
edge_connection("Kuningan",['Cirebon'])

nx.draw(lingko, with_labels=True)
plt.show()
```
